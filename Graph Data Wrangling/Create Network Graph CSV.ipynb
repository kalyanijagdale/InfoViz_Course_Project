```python
import pandas
import json
```

```python
data = pandas.read_csv("tmdb_5000_credits.csv")
data = data[:200]
```

```python
print(data['cast'])
```

```python
cast_filtered = []
```

```python
for item in data['cast']:
    cast_curr = []
    item = item.split('{')
    item = item[1:]
    for member in item:
        cast_curr.append(member.split("name")[1].split(":")[1].split(",")[0].strip('"')[2:])
    #for cast_member in item:
        #item = json.load(cast_member)
        #print(cast_member)
        #cast_curr.append(item['name'])
    cast_filtered.append(cast_curr)
```

```python
temp = []
for item in cast_filtered:
    temp.append(item[:4])
cast_filtered = temp
```

```python
print(len(cast_filtered))
```

```python
edges = []
```

```python
for cast in cast_filtered:
    for i in range(len(cast)):
        curr = cast[i]
        for j in range(len(cast)):
            if curr != cast[j]:
                if [curr, cast[j]] not in edges and [cast[j], curr] not in edges:
                    edges.append([curr, cast[j]])
```

```python
edges
```

```python
csv = "source,target,value\n"
```

```python
for item in edges:
    row = item[0] + "," + item[1] + "," + "0\n"
    csv += row
```
